### Generación y adecuación de datos sintéticos
* Generación de datos
* Visualización
* Codificación de la salida en one-hot
* Separación de los conjuntos de training y test

Todo utilizando la librería ScikitLearn


```python
#Importamos las librerías necesarias, tensor y scikitlearn
import tensorflow as tf
from sklearn import datasets as skds
```

```python
#Generación de datos
#Para la generación de datos utilizamos el comando skds.make_classification() al que hay que pasarle 
#el número de muestras, el número de características (2) el número de clases de salida(2)y el número 
#de clauster por clase y devuelve dos array, X con las muestras e y con las etiquetas de cada entrada

X, y = skds.make_classification(n_samples=400,
   n_features=2,
   n_informative=2, 
   n_redundant=0,  #caracteristicas redundantes
   n_repeated=0, #Tuplas duplicadas
   n_classes=2,
   n_clusters_per_class=1)
#Podemos acceder a su contenido
print(X[1:3,])
print(y[1:5])
```

```python
# Visualización
#Para este punto importamos la clase matplotlib y dibujamos las muestras por clase
import matplotlib.pyplot as plt
plt.scatter(X[:,0],X[:,1],marker='o',c=y)
plt.show()
```

```python
#Codificación de la variable salida
#Convertimos la salida en one hot o one-hot encoded
#Hablar de las variables dummies. Para ello vamos a usar la función eye de numpy
import numpy as np
print(y[1:6])

#Indicamos a eye que convierta y en una variable de 2 variables
y=np.eye(2)[y]
print(y[1:6])
```

```python
#Separación del conjunto de datos
#Ahora necesitamos particionar nuestros datos entre training y test. Para eso usamos también la librería sklearn
from sklearn.model_selection import train_test_split
#para dividir el conjunto de datos usamos train_test_split, pasándole las muestras, las etiquetas,
#el tamaño por donde queremos partir,
#devuelve 4 arrays: las muestras de entrada para los conjuntos de train y test, y las muestras de salida para train y test

#Metodo     train_test_split
    
X_train, X_test, y_train, y_test = (train_test_split(X, y,test_size=0.20, random_state=17))
```
